```python
import numpy as np
from scipy.stats import hypergeom
from scipy.stats import fisher_exact
```

```python
# collusion_dynamic_adversarial-both
s1_ped = [1,	1,	0,	5,	1,	2,	1,	2,	2,	1]
s1_stop = [26,	19,	30,	25,	18,	23,	24,	26,	17,	15]
# single_dynamic_adversarial
s2_ped = [5,	2,	5,	5,	7,	3,	7,	1,	4,	3]
s2_stop = [20,	11,	13,	8,	9,	10,	12,	12,	8,	11]
table = np.array([[np.sum(s1_stop), np.sum(s1_ped)], [np.sum(s2_stop), np.sum(s2_ped)]])
odds_ratio, p_value = fisher_exact(table)
print(f"Odds Ratio: {odds_ratio:.4f}")
print(f"P-value: {p_value:.4e}")
```
