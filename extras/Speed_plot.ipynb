```python
import pandas as pd

# Define the file paths
static_csv_path = '/content/velocity_collusion_static_adversarial-both.csv'
dynamic_csv_path = '/content/velocity_collusion_dynamic_adversarial-both.csv'

# Read the CSV files into pandas DataFrames
df_static = pd.read_csv(static_csv_path)
df_dynamic = pd.read_csv(dynamic_csv_path)

# Convert DataFrames to NumPy arrays
static_data_array = df_static.to_numpy()
dynamic_data_array = df_dynamic.to_numpy()

# Print the first few rows of each array to verify
print("Static data array (first 5 rows):")
print(static_data_array[:5])
print("\nDynamic data array (first 5 rows):")
print(dynamic_data_array[:5])
```

```python
import matplotlib.pyplot as plt

# Extract data from the 5th to the 120th element (Python indexing: 4 to 119)
static_plot_data = static_data_array[15:121]
dynamic_plot_data = dynamic_data_array[4:110]

# Separate x and y values for plotting
x_static = static_plot_data[:, 0]
y_static = static_plot_data[:, 1]

x_dynamic = dynamic_plot_data[:, 0]
y_dynamic = dynamic_plot_data[:, 1]

# --- Modify to shift static plot to the left ---
# Define a shift value (e.g., 2 to shift left by 2 units on the x-axis)
shift_value_static = 11
x_static_shifted = x_static - shift_value_static
# ------------------------------------------------

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(x_static_shifted, y_static, label='static', marker='o', markersize=3, linestyle='-') # Use shifted x_static
plt.plot(x_dynamic, y_dynamic, label='dynamic', marker='x', markersize=3, linestyle='--')

# Add title and labels with bigger fonts
plt.title('Speed comparison between dynamic and static scenarios in the collusion attack', fontsize=16)
plt.xlabel('Frame', fontsize=14)
plt.ylabel('Speed (m/s)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Save the plot as a PDF for highest quality
plt.savefig('speed_comparison.pdf', format='pdf', dpi=600)

plt.show()
```
